```python
import dlib
from google.colab.patches import cv2_imshow
```

```python
image = cv2.imread('/content/drive/MyDrive/Computer Vision Masterclass - Udemy/Images/people2.jpg')
cv2_imshow(image)
```

```python
face_detector_hog = dlib.get_frontal_face_detector()
```

```python
detections = face_detector_hog(image, 1) #Higher value gets smaller size of bounding box (We have used 1 there)
```

```python
detections , len(detections)
```

```python
for face in detections:
  #print(face)
  #print(face.left())
  #print(face.top())
  #print(face.right())
  #print(face.bottom())
  l, t, r, b = face.left(), face.top(), face.right(), face.bottom()
  cv2.rectangle(image, (l, t), (r, b), color = (0,0,255), thickness = 1)

cv2_imshow(image)
```
