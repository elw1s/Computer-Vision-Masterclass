```python
import dlib
from google.colab.patches import cv2_imshow
```

```python
image = cv2.imread('/content/drive/MyDrive/Computer Vision Masterclass - Udemy/Images/people2.jpg')
cv2_imshow(image)
```

```python
cnn_detector = dlib.cnn_face_detection_model_v1('/content/drive/MyDrive/Computer Vision Masterclass - Udemy/Weights/mmod_human_face_detector.dat')
```

```python
detections = cnn_detector(image, 1) #The parameter 1 is indicating the scale of the image
```

```python
detections, len(detections)
```

```python
for face in detections:
  l, t, r, b, c = face.rect.left(), face.rect.top(), face.rect.right(), face.rect.bottom(), face.confidence
  print(c) #Confidence score that makes sure about faces
  cv2.rectangle(image, (l,t), (r,b), color = (0, 0, 255), thickness = 1)
```
